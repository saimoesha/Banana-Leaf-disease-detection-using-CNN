```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from keras.utils import to_categorical
from keras.optimizers import Adam
import pandas as pd

# Parameters
img_size = 64  # Resize images to 64x64
num_classes = 3  # Banana Bacterial Wilt, Black Sigatoka Disease, Healthy

# Directory containing images
base_dir = "/content/drive/MyDrive/OriginalSet"  # Replace with your dataset path
categories = ['Banana Bacterial Wilt','Black sigatoka disease','Healthy']


# Data Preparation
def load_data():
    X, y = [], []
    for idx, category in enumerate(categories):
        category_path = os.path.join(base_dir, category)
        for img_name in os.listdir(category_path):
            img_path = os.path.join(category_path, img_name)
            try:
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Read in grayscale
                img = cv2.resize(img, (img_size, img_size))  # Resize
                X.append(img)
                y.append(idx)  # Class label
            except Exception as e:
                print(f"Error loading image {img_path}: {e}")
    return np.array(X), np.array(y)

# Load dataset
X, y = load_data()

# Normalize and reshape
X = X / 255.0  # Normalize pixel values
X = X.reshape(-1, img_size, img_size, 1)  # Add channel dimension
y_one_hot = to_categorical(y, num_classes)  # One-hot encode labels

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)

# ANN Model
ann_model = Sequential([
    Flatten(input_shape=(img_size, img_size, 1)),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])

ann_model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Train ANN
ann_history = ann_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=16)

# CNN Model
cnn_model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(img_size, img_size, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.3),
    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

cnn_model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Train CNN
cnn_history = cnn_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=16)

# Evaluate ANN and CNN
ann_eval = ann_model.evaluate(X_test, y_test, verbose=0)
cnn_eval = cnn_model.evaluate(X_test, y_test, verbose=0)

# Predict and generate reports
ann_preds = np.argmax(ann_model.predict(X_test), axis=1)
cnn_preds = np.argmax(cnn_model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)

# ANN Classification Report
print("ANN Classification Report:")
print(classification_report(y_true, ann_preds, target_names=categories))

# CNN Classification Report
print("CNN Classification Report:")
print(classification_report(y_true, cnn_preds, target_names=categories))

# Display Results Table
results = pd.DataFrame({
    'Metric': ['Accuracy', 'Loss', 'Val Accuracy', 'Val Loss'],
    'ANN': [ann_history.history['accuracy'][-1],
            ann_history.history['loss'][-1],
            ann_history.history['val_accuracy'][-1],
            ann_history.history['val_loss'][-1]],
    'CNN': [cnn_history.history['accuracy'][-1],
            cnn_history.history['loss'][-1],
            cnn_history.history['val_accuracy'][-1],
            cnn_history.history['val_loss'][-1]]
})
print("\nModel Performance Comparison:")
print(results)

# Visualize Predictions for Each Class
def visualize_predictions(model, X, y_true, categories):
    fig, axes = plt.subplots(1, len(categories), figsize=(15, 5))
    for i, category in enumerate(categories):
        idx = np.where(np.argmax(y_true, axis=1) == i)[0][0]
        pred = model.predict(X[idx:idx+1])
        predicted_class = categories[np.argmax(pred)]
        true_class = categories[i]
        axes[i].imshow(X[idx].reshape(img_size, img_size), cmap='gray')
        axes[i].set_title(f"True: {true_class}\nPred: {predicted_class}")
        axes[i].axis('off')

# Display for ANN
print("\nANN Predictions:")
visualize_predictions(ann_model, X_test, y_test, categories)

# Display for CNN
print("\nCNN Predictions:")
visualize_predictions(cnn_model, X_test, y_test, categories)

```

```python
# Visualize predictions with True and Predicted labels for both CNN and ANN for 9 images

def visualize_predictions(model, X, y_true, categories, model_name="Model"):
    # Setup figure with size to display 9 images (3 rows, 3 columns)
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))

    # Iterate through each class and show images for each
    for i, category in enumerate(categories):
        idx = np.where(np.argmax(y_true, axis=1) == i)[0]  # Get indices for current category
        selected_idx = np.random.choice(idx, 3)  # Randomly select 3 images from the class

        for j, idx in enumerate(selected_idx):
            sample_img = X[idx]

            # Make predictions using the model
            pred = model.predict(sample_img.reshape(1, img_size, img_size, 1))
            predicted_class = categories[np.argmax(pred)]
            true_class = categories[i]

            # Plot the image
            axes[i, j].imshow(sample_img.reshape(img_size, img_size), cmap='viridis')  # Using 'viridis' for colored images
            axes[i, j].set_title(f"True: {true_class}\nPred: {predicted_class}")
            axes[i, j].axis('off')  # Hide axes for better visuals

    fig.suptitle(f"{model_name} - Predictions (9 Images)", fontsize=16)
    plt.tight_layout()
    plt.show()

# Display for ANN
print("\nANN Predictions for 9 Images:")
visualize_predictions(ann_model, X_test, y_test, categories, model_name="ANN")

# Display for CNN
print("\nCNN Predictions for 9 Images:")
visualize_predictions(cnn_model, X_test, y_test, categories, model_name="CNN")
```

```python
import pandas as pd

# Function to get the metrics (accuracy, loss, val_accuracy, val_loss) for a given model
def get_model_metrics(model_history, model_name="Model"):
    metrics = {
        "Metric": ["Training Accuracy", "Validation Accuracy", "Training Loss", "Validation Loss"],
        model_name: [
            model_history.history['accuracy'][-1],  # Last training accuracy
            model_history.history['val_accuracy'][-1],  # Last validation accuracy
            model_history.history['loss'][-1],  # Last training loss
            model_history.history['val_loss'][-1]  # Last validation loss
        ]
    }
    return pd.DataFrame(metrics)

# Get metrics for ANN model
ann_metrics = get_model_metrics(ann_history, "ANN")

# Get metrics for CNN model
cnn_metrics = get_model_metrics(cnn_history, "CNN")

# Merge both ANN and CNN metrics into a single DataFrame for easy comparison
metrics_comparison = pd.concat([ann_metrics.set_index('Metric'), cnn_metrics.set_index('Metric')], axis=1)

# Display the table
print("\nModel Performance Comparison Table:")
print(metrics_comparison)
```
